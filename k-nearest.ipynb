```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import joblib


```

```python
# Load Data
df = pd.read_csv(r'C:/Users/Asus/Desktop/dl_lab/heart_ml_lvl2/heart.csv')
print("Original Data Loaded:", df.shape)

# Data Cleaning
df = df.drop_duplicates()
df = df.fillna(df.median(numeric_only=True))
df = df.fillna(df.mode().iloc[0])
print("After Cleaning:", df.shape)

# Pre-processing
for c in df.select_dtypes('object'):
    df[c] = LabelEncoder().fit_transform(df[c])

print("After Encoding:")
print(df.head())

# Feature Selection
corr = df.corr()['target'].abs()
features = corr[corr > 0.1].index.drop('target')

print("Co-relations:\n", corr)
print("\nSelected Features:", list(features))

# Split Data
x = df[features]
y = df['target']

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)

# K-Nearest Neighbors Model
model = KNeighborsClassifier(n_neighbors=5)

# Train Model
model.fit(x_train, y_train)

# Predict
y_pred = model.predict(x_test)

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: {:.2f}%".format(accuracy * 100))

# Save Model
joblib.dump(model, r'C:/Users/Asus/Desktop/dl_lab/heart_ml_lvl2/heart_knn_model.pkl')
print("heart_knn_model.pkl file created successfully!")
```
